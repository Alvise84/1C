# Restaurant Management System

## **Описание проекта:**

"Restaurant Management System" — это комплексное решение 1C для автоматизации управления рестораном, включающее интеграцию с Telegram для бронирования столиков, использование нейросети для генерации описаний блюд, а также различные улучшения для повышения эффективности работы персонала и удовлетворенности клиентов. Проект направлен на оптимизацию процессов бронирования, улучшение взаимодействия с клиентами и анализ данных для принятия обоснованных управленческих решений.

## Этапы проекта:

1. Создание и настройка Telegram-бота
2. Интеграция с нейросетью ChatGPT
3. Разработка функционала для удаления брони
4. Интеграция с программой лояльности
5. Настройка сообщений об ошибках
6. Добавление поля для комментариев
7. Автоматическая отправка сообщений гостю
8. Создание отчёта "Пик загруженности"
